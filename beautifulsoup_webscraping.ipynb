```python
from bs4 import BeautifulSoup as bs
import requests
```

```python
#storing data from website 
r = requests.get("https://en.wikipedia.org/wiki/South_Africa")

soup = bs(r.content)
contents = soup.prettify()
print(contents)
```

```python
#selecting specific rows.
info_box = soup.find(class_="infobox-above")
info_rows = info_box.find_all("tr")
for row in info_rows:
    print(row.prettify())
```

```python
 #creating an empty dictionery and store table block values
#clean up dictionary using string replacement

def get_content_value(row_data):
    if row_data.find("li"):
        return[li.get_text().replace("\xa0", "") for li in row.find_all("li")]
    else:
        return row_data.get_text().replace("\xa0", "")
country_info =  {}

for index, row in enumerate(info_rows):
    if index == 0:
        country_info['title'] = row.find("th").get_text()
    elif index == 1:
        continue
    else:
        content_key = row.find("th").get_text()
        content_value = get_content_value(row.find("td"))
        country_info[content_key] = content_value
        
country_info
```
